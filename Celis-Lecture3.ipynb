$\textbf{Directed Graph}$

 In simple terms, a Directed Graph or a Digraph for short is a graph where each edge has a direction. For example, let us imagine the digraph as a sort of map, where we can only go in one direction from vertex to vertex. In the graph below we can walk from vertex 1 to 2 but not necessarily from 2 to 1. 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_nodes_from([1, 2, 3, 4, 5])
G.add_edges_from([(1,2), (2,3), (3, 4), (4, 5), (5,1)])
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels = True)
plt.show()
```

$\textbf{Simple Graphs}$

A simple graph is simply a graph that has no loops or parallel edges. Simple graphs are also undirected as opposed to directed graphs. The example below has no parallel edges, no loops, and is undirected therefore it is a simple graph. 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph() 
G.add_nodes_from([1,2,3])
G.add_edges_from([(1,2), (2,3), (3,1)])
nx.draw(G, with_labels = True)
plt.show()
```

$\textbf{Complete Graphs}$

Complete graphs are undirected graphs where each edge is connected to every distinct pair of vertices. Simply put, 
a complete graph is a graph where each vertex is connected to every other vertex. 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.complete_graph(5)
nx.draw(G, with_labels = True)
plt.show()
```

$\textbf{Bipartite Graph}$

A bipartite graph is a graph wherein vertices can be divided into two disjoint sets such that no two vertices in the same set are adjacent. Every vertex in one set is connected to a vertex in the other set. 

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
G.add_nodes_from([1, 2, 3], bipartite=0) 
G.add_nodes_from(['A', 'B', 'C'], bipartite=1)
G.add_edges_from([(1, 'B'), (1, 'C'), (2, 'B'), (2, 'A'), (3, 'C')])
pos = nx.bipartite_layout(G, nodes=[1, 2, 3])  
nx.draw(G, pos, with_labels=True)
plt.show()
```

$\textbf{Subgraphs}$

In simple terms, a subgraph is a graph that is formed from a subset of the vertices and edges of another graph. 

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4)])

# Original Graph 
sub_nodes = [1, 2, 3]
subgraph = G.subgraph(sub_nodes)
plt.figure(figsize=(10, 5))
plt.subplot(121)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='black')
plt.title('Parent Graph')

#Subgraph
plt.subplot(122)
nx.draw(subgraph, with_labels=True, node_color='red', edge_color='black')
plt.title('Subgraph')
plt.tight_layout()
plt.show()
```

$\textbf{Degree of a Vertex}$

The degree of a vertex is the number of edges incident on a particular vertex. It is essentially how many edges are connected to that particular vertex. In the example below, each vertex has a degree of two since each one of them has
two edges connected to them. 

deg(1) = 2

deg(2) = 2

deg(3) = 2

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from([1, 2, 3])
G.add_edges_from([(1, 2), (1, 3), (2, 3)])
nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Walk}$

In graph theory, a Walk is a sequence of vertices and edges in a graph such that each edge in the sequence connects the preceding vertex to the succeeding vertex. In the example below, a walk could be from $A \to C \to D \to E \to C$.



```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D','E','F'])
G.add_edges_from([('A','B'), ('A','C'), ('B','D'), ('C','D'), ('C','E'), ('D','E'), ('E','F')])
nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Trail}$

Similar to a walk, a trail is a walk with no repeating edges, however vertices can be repeated. We can take the example above again, and create a trail. A trail could be $A \to B \to D \to C \to E \to D$. Notice how we have no repeating edges, therefore this is a trail. 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D','E','F'])
G.add_edges_from([('A','B'), ('A','C'), ('B','D'), ('C','D'), ('C','E'), ('D','E'), ('E','F')])
nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Path}$

A path is simply a trail with no repeating vertices. Since it has no repeating vertex, as a result no edge will also repeat similar to a trail. In the same graph, an example of a path would be $F \to E \to D \to B \to A \to C$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D','E','F'])
G.add_edges_from([('A','B'), ('A','C'), ('B','D'), ('C','D'), ('C','E'), ('D','E'), ('E','F')])
nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Closed Walk}$

A closed walk is simply when a walk starts and ends in the same vertex. In a closed walk, the vertices and edges can be repeated. In the example, we can create a closed walk by doing $A \to B \to D \to C \to A$. Notice how we start and end in the same vertex. 


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D'])
G.add_edges_from([('A','B'), ('A','C'), ('B','D'), ('C','D'), ('C', 'B')])
nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Circuit}$

A circuit is also a closed walk but there are no repeating edges. It is like if a closed walk and trail had a baby. Repeating vertices are allowed. In the example below, we can create a circuit by $B \to C \to A \to B$. 


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D'])
G.add_edges_from([('A','B'), ('A','C'), ('C','D'), ('C', 'B')])
nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Simple Circuits}$

A simple circuit is a circuit where all vertices are distinct except for the starting and ending vertex. This means that a simple circuit does not repeat any vertices or edges, except for the first and last vertices. In the example below, we can create a simple circuit by doing $D \to E \to A \to E \to B \to C \to D$
 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D', 'E'])
G.add_edges_from([('A','B'), ('A','C'), ('C','D'), ('C', 'B'), ("D", "E"), ("E", "A")])
nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Connected Graphs}$

A connected graph is simpley a graph wherein there is a path from one vertex to every other vertex. A very easy example is simply a straight line graph. In this example, we can can see that each vertex is connected to all other vertices either directly or indirectly. 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D', 'E'])
G.add_edges_from([('A','B'), ('B','C'), ('C','D'), ('D', 'E')])
nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Euler's Circuit}$

An Euler's Circuit is a circuit which traverses every edge exactly once and returns to the starting vertex. In the example below we can create an Euler's Circuit by simply doing $A \to C \to D \to B \to A$. 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D'])
G.add_edges_from([('A','B'), ('A','C'), ('B','D'), ('C','D')])
nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Euler's Trail}$

Similar to an Euler's Circuit, an Euler's Trail traverses each edge exactly once, but may not end in the same vertex. In this example we can create an Euler's Trail by doing $A \to E \to D \to C \to A \to B \to C$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D', 'E'])
G.add_edges_from([('A','B'), ('A','C'), ('C','D'), ('C', 'B'), ("D", "E"), ("E", "A")])
nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Hamiltonian Circuit}$

A Hamiltonian Circuit is somewhat the opposite of an Euler's Circuit. Instead of traversing each edge, we traverse each vertex exactly once and end on the same starting vertex. In this example, we can see that that we can create a Hamiltonian Circuit by doing the following $A \to B \to D \to C \to A.$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D'])
G.add_edges_from([('A','B'), ('A','C'), ('B','D'), ('C','D'), ('C', 'B')])
nx.draw(G, with_labels=True)
plt.show() 
```
