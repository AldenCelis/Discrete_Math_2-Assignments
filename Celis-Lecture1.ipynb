To prove that Euler's number, $( e )$, is irrational, we start by assuming the opposite, that is, $( e )$ can be expressed as a fraction $( \frac{p}{q} )$, where $( p )$ and $( q )$are integers with no common factors other than 1, and $( q \neq 0 )$.

We recall the definition of $( e )$ as the sum of the infinite series:

$[
e = \sum_{k=0}^\infty \frac{1}{k!}
]$

Let's denote the $( n )$th partial sum of this series as $( S_n ):$

$[
S_n = \sum_{k=0}^n \frac{1}{k!}
]$

Since we assumed $( e )$ is rational, $( S_n )$ must approach a rational limit as $( n)$ goes to infinity. However, since each term in the series $( \frac{1}{k!} )$ is positive, $( S_n )$ is increasing.

We know that for any $( k \geq 1 )$, $( k! > k )$, which implies $( \frac{1}{k!} < \frac{1}{k} )$. This means that the series $( \sum_{k=0}^\infty \frac{1}{k!} )$ converges.

Now, if $( S_n )$ converges to a rational number, it must be bounded. However, this contradicts the divergence of the harmonic series, which is made of larger terms.

Hence, assuming $( e )$ is rational leads to a contradiction. Therefore, we conclude that $( e )$ is not rational, i.e., $( e \notin \mathbb{Q} )$.

We aim to prove Minkowski's Inequality for sums of positive real numbers $( (a_k, b_k) > 0 )$ and $( p > 1 )$:

$
\begin{equation}
\left( \sum_{k=1}^n |a_k + b_k|^p \right)^\frac{1}{p} \leq \left( \sum_{k=1}^n |a_k|^p \right)^\frac{1}{p} + \left( \sum_{k=1}^n |b_k|^p \right)^\frac{1}{p}
\end{equation}
$
$
\\ \textbf{Step 1: Power-of-a-Power Rule}
$
We know that $( (n^p)^\frac{1}{p} = n )$ for any base number $( n )$. Applying this to the sums of absolute values raised to the power of $( p )$, we can rewrite the inequality:

$
\begin{equation}
\sum_{k=1}^n |a_k + b_k| \leq \sum_{k=1}^n |a_k| + \sum_{k=1}^n |b_k|
\end{equation}
$

$\\ \textbf{Step 2: Property of Summation}$

The summation operation $( \sum )$ distributes over addition. Therefore:

$
\begin{equation}
\sum_{k=1}^n |a_k + b_k| \leq \sum_{k=1}^n |a_k| + \sum_{k=1}^n |b_k|
\end{equation}
$
$\\ \textbf{Step 3: Triangle Inequality}$

For clarity, let's remove the summation notation:

$[ |a_k + b_k| \leq |a_k| + |b_k| ]$

Since $( (a_k, b_k) > 0 )$, both $( a_k )$ and $( b_k )$ are positive real numbers. This guarantees that $( (a_k + b_k) )$ is also positive, and the absolute value operation doesn't change the sign. Therefore, $( |a_k + b_k| = a_k + b_k )$. Similarly, $( |a_k| = a_k )$ and $( |b_k| = b_k )$.

Substituting these equalities, we get:

$[ a_k + b_k \leq a_k + b_k ]$

This is trivially true, demonstrating that the inequality holds for any positive real numbers $( a_k )$ and $( b_k )$.

$\\ \textbf{Conclusion}$

By utilizing the power-of-a-power rule, the property of summation, and the triangle inequality, we have proven Minkowski's Inequality for sums of positive real numbers with $( p > 1 )$.

4. Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

We will prove Sedrakayan's Lemma for non-negative real numbers $(u_i, v_i) \in \mathbb{R}^+$:

$$\frac{\left( \sum\limits_{i=1}^n u_i \right)^2}{\sum\limits_{i=1}^n v_i} \leq \sum\limits_{i=1}^n \frac{u_i^2}{v_i}$$

$\textbf{Step 1}$
the Cauchy-Schwarz Inequality

$$\left( \frac{\sum\limits_{i=1}^n u_i}{\sum\limits_{i=1}^n v_i} \right)^2 \leq \frac{\sum\limits_{i=1}^n u_i^2}{v_i}$$

$\textbf{Step 2}$

The Cauchy-Schwarz Inequality states that for any real numbers $(a_i)$ and $(b_i)$:

$$\left( \sum\limits_{i=1}^n a_i b_i \right)^2 \leq \left( \sum\limits_{i=1}^n a_i^2 \right) \left( \sum\limits_{i=1}^n b_i^2 \right)$$


$\textbf{Step 3}$

Let's define:

* $a_i = \sqrt{u_i}$
* $b_i = \frac{1}{\sqrt{v_i}}$

$$\left( \sum\limits_{i=1}^n \sqrt{u_i} \cdot \frac{1}{\sqrt{v_i}} \right)^2 \leq \left( \sum\limits_{i=1}^n u_i \right) \left( \sum\limits_{i=1}^n \frac{1}{v_i} \right)$$

$\textbf{Step 4}$

* **Left-hand side:**

$$\left( \sum\limits_{i=1}^n \sqrt{u_i} \cdot \frac{1}{\sqrt{v_i}} \right)^2 = \left( \frac{\sum\limits_{i=1}^n u_i}{\sum\limits_{i=1}^n v_i} \right)^2$$


* **Right-hand side:**

$$\left( \sum\limits_{i=1}^n u_i \right) \left( \sum\limits_{i=1}^n \frac{1}{v_i} \right) = \frac{\sum\limits_{i=1}^n u_i^2}{v_i}$$


$\textbf{Step 5}$

Since the reformulated inequality on the left-hand side is less than or equal to the right-hand side, as derived from the Cauchy-Schwarz Inequality, we have successfully demonstrated the validity of Sedrakayan's Lemma for all non-negative real numbers.$(u_i, v_i) \in \mathbb{R}^+$.
